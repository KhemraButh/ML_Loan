```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

    # Your GeoDataFrame (shown in the output)
cambodia = gpd.GeoDataFrame.from_file("/Users/thekhemfee/Downloads/Intelligent_Loan_Approval/frontend/geoBoundaries-KHM-ADM1_simplified.geojson")  # or from your source

    # Your disbursement data
disbursements = pd.DataFrame({
        'Province': ['Stung Treng', 'Phnom Penh', 'Kampong Thom', 'Koh Kong'],
        'Amount': [500000, 1200000, 750000, 300000]
    })

    # 1. Merge data
merged = cambodia.merge(disbursements,
                        left_on='shapeName',
                        right_on='Province',
                        how='left')

    # 2. Plot with matplotlib
fig, ax = plt.subplots(figsize=(6, 4))
merged.plot(column='Amount',
                ax=ax,
                legend=True,
                cmap='YlOrRd',
                missing_kwds={'color': 'lightgrey'},
                edgecolor='black')
plt.title("CMCB Disbursements by Province")
plt.axis('off')

    # 3. Add labels (optional)
for idx, row in merged.iterrows():
  plt.annotate(text=f"{row['shapeName']}\n${row['Amount']:,}",
                    xy=row['geometry'].centroid.coords[0],
                    ha='center',
                    fontsize=4)
```
